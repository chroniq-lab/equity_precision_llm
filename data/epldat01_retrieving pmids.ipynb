```python
import pandas as pd
from pymed import PubMed
from metapub import PubMedFetcher 
```

```python
concept1 = '"precision medicine"[Title/Abstract] OR "personalized medicine"[Title/Abstract] OR "stratified medicine"[Title/Abstract] OR "individualized medicine"[Title/Abstract] OR "computational biology"[Title/Abstract] OR “computational biology”[MeSH Major Topic] OR genom*[Title/Abstract] OR genom*[MeSH Major Topic] OR transcript*[Title/Abstract] OR transcript*[MeSH Major Topic] OR proteom*[Title/Abstract] OR proteom*[MeSH Major Topic] OR lipid*[Title/Abstract] OR lipid*[MeSH Major Topic] OR metagenom*[Title/Abstract] OR metagenom*[MeSH Major Topic] OR exposom*[Title/Abstract] OR exposom*[MeSH Major Topic] OR glycom*[Title/Abstract] OR glycom*[MeSH Major Topic] OR pharmacogenom*[Title/Abstract] OR pharmacogenom*[MeSH Major Topic] OR metabolom*[Title/Abstract] OR metabolom*[MeSH Major Topic] '

concept2 = '"diabetes mellitus, type 2"[Title/Abstract] OR "diabetes mellitus, type 2"[MeSH Major Topic] OR diabetes[Title/Abstract] OR diabetes[MeSH Major Topic] OR glucose[Title/Abstract] OR glucose[MeSH Major Topic] OR diabetic*[Title/Abstract] OR diabetic*[MeSH Major Topic]'

years = '("2010"[Date - Publication] : "2023"[Date - Publication])'

final_query = "(" + concept1 + ") AND (" + concept2 + ") AND " + years +")"
```

https://stackoverflow.com/questions/68952738/get-full-paper-content-from-pubmed-via-api-and-list-of-ids
https://pypi.org/project/paperscraper/

```python

pubmed = PubMed(tool="PubMedSearcher", email="jvargh7@emory.edu")

results = pubmed.query(final_query, max_results=500)


```

```python
articleList = []
articleInfo = []
pmid_list = []

for article in results:
# Print the type of object we've found (can be either PubMedBookArticle or PubMedArticle).
# We need to convert it to dictionary with available function
    articleDict = article.toDict()
    articleList.append(articleDict)

```

```python
articleList[0]['keywords']
```

```python
# Generate list of dict records which will hold all article details that could be fetch from PUBMED API
for article in articleList:
#Sometimes article['pubmed_id'] contains list separated with comma - take first pubmedId in that list - thats article pubmedId
    pmid_list.append(article['pubmed_id'].partition('\n')[0])

print(pmid_list)
```

```python
fetch = PubMedFetcher()

abstracts_list = []

for pmid in pmid_list:
    article = fetch.article_by_pmid(pmid)
    print(article.abstract)
```
